```python
# --- TAHAP 1: PERSIAPAN DATA ---

# 1) Import library yang kita butuhin buat tugas ini
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2) Membaca file excel yang udah di-upload
# Pastikan di dalam tanda petik ini adalah hasil paste dari "Salin jalur" file Excel-mu
df = pd.read_excel('/content/UAS Python.xlsx')

# 3) Tampilkan 6 data pertama buat ngecek apakah tabelnya udah bener
df.head(6)
```

```python
# 4) Tampilkan 6 data terakhir buat ngeliat bagian bawah tabel
df.tail(6)
```

```python
# 5) Cek struktur data sama tipe data tiap kolomnya
df.info()
```

```python
# 6) Tampilkan ringkasan statistik awal buat kolom yang isinya angka
df.describe()
```

```python
# 7) Tampilkan jumlah total baris sama kolom di dataset ini
df.shape
```

```python
# --- TAHAP 2: ANALISIS STATISTIK DESKRIPTIF DASAR ---

# 1) Hitung rata-rata penjualan dari kolom Sales
rata_sales = df['Sales'].mean()
print("Rata-rata penjualan:", rata_sales)
```

```python
# 2) Hitung nilai statistik deskriptif lainnya buat kolom Sales
print("Penjualan Minimum :", df['Sales'].min())
print("Penjualan Maksimum :", df['Sales'].max())
print("Median Penjualan   :", df['Sales'].median())
print("Standar Deviasi    :", df['Sales'].std())
print("Kuartil Pertama(Q1):", df['Sales'].quantile(0.25))
print("Kuartil Ketiga (Q3):", df['Sales'].quantile(0.75))
```

```python
# 3) Tampilkan Top 10 Order ID dengan penjualan terbesar
top10_terbesar = df.nlargest(10, 'Sales')[['Order ID', 'Sales']]
print("--- TOP 10 PENJUALAN TERBESAR ---")
print(top10_terbesar)
```

```python
# 4) Tampilkan Top 10 Order ID dengan penjualan terkecil
top10_terkecil = df.nsmallest(10, 'Sales')[['Order ID', 'Sales']]
print("--- TOP 10 PENJUALAN TERKECIL ---")
print(top10_terkecil)
```

```python
# 5) Hitung jumlah transaksi dari masing-masing kategori

print("--- JUMLAH TRANSAKSI BERDASARKAN SHIP MODE ---")
print(df['Ship Mode'].value_counts())
print("\n")

print("--- JUMLAH TRANSAKSI BERDASARKAN SEGMENT ---")
print(df['Segment'].value_counts())
print("\n")

print("--- JUMLAH TRANSAKSI BERDASARKAN REGION ---")
print(df['Region'].value_counts())
print("\n")

print("--- JUMLAH TRANSAKSI BERDASARKAN CATEGORY ---")
print(df['Category'].value_counts())
print("\n")

print("--- JUMLAH TRANSAKSI BERDASARKAN SUB-CATEGORY ---")
print(df['Sub-Category'].value_counts())
print("\n")

print("--- JUMLAH TRANSAKSI BERDASARKAN ORDER PRIORITY ---")
print(df['Order Priority'].value_counts())
```

```python
# 6) Hitung total nilai penjualan (Sales) buat masing-masing kategori

print("--- TOTAL PENJUALAN BERDASARKAN SHIP MODE ---")
print(df.groupby('Ship Mode')['Sales'].sum())
print("\n")

print("--- TOTAL PENJUALAN BERDASARKAN SEGMENT ---")
print(df.groupby('Segment')['Sales'].sum())
print("\n")

print("--- TOTAL PENJUALAN BERDASARKAN REGION ---")
print(df.groupby('Region')['Sales'].sum())
print("\n")

print("--- TOTAL PENJUALAN BERDASARKAN CATEGORY ---")
print(df.groupby('Category')['Sales'].sum())
print("\n")

print("--- TOTAL PENJUALAN BERDASARKAN SUB-CATEGORY ---")
print(df.groupby('Sub-Category')['Sales'].sum())
print("\n")

print("--- TOTAL PENJUALAN BERDASARKAN ORDER PRIORITY ---")
print(df.groupby('Order Priority')['Sales'].sum())
```

```python
# --- TAHAP 3: ANALISIS PENJUALAN BERDASARKAN LOKASI ---

# 1) Tampilkan total penjualan berdasarkan Market
print("--- TOTAL PENJUALAN BERDASARKAN MARKET ---")
print(df.groupby('Market')['Sales'].sum())
print("\n")

# 2) Tampilkan total penjualan berdasarkan Region
print("--- TOTAL PENJUALAN BERDASARKAN REGION ---")
print(df.groupby('Region')['Sales'].sum())
```

```python
# 3) Tampilkan 5 Negara (Country) dengan total penjualan terbesar
top5_negara = df.groupby('Country')['Sales'].sum().nlargest(5)
print("--- TOP 5 NEGARA PENJUALAN TERBESAR ---")
print(top5_negara)
print("\n")

# 4) Tampilkan 5 Kota (City) dengan total penjualan terbesar
top5_kota = df.groupby('City')['Sales'].sum().nlargest(5)
print("--- TOP 5 KOTA PENJUALAN TERBESAR ---")
print(top5_kota)
```

```python
# 1) Ambil data 5 kota dengan penjualan terbesar
top5_kota = df.groupby('City')['Sales'].sum().nlargest(5).reset_index()
print("--- TOP 5 KOTA PENJUALAN TERBESAR ---")
print(top5_kota)

# 2) Bikin grafik batang buat top 5 kota
plt.figure(figsize=(8, 4))
sns.barplot(data=top5_kota, x='City', y='Sales', palette='Blues_r')
plt.title('Top 5 Kota dengan Total Penjualan Terbesar')
plt.xlabel('Kota')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
# 3) Ambil data 5 state/provinsi dengan penjualan terbesar
top5_state = df.groupby('State')['Sales'].sum().nlargest(5).reset_index()
print("--- TOP 5 STATE PENJUALAN TERBESAR ---")
print(top5_state)

# 4) Bikin grafik batang buat top 5 state
plt.figure(figsize=(8, 4))
sns.barplot(data=top5_state, x='State', y='Sales', palette='Oranges_r')
plt.title('Top 5 State/Provinsi dengan Total Penjualan Terbesar')
plt.xlabel('State / Provinsi')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
# 5) Ambil data 5 region dengan penjualan terbesar
top5_region = df.groupby('Region')['Sales'].sum().nlargest(5).reset_index()
print("--- TOP 5 REGION PENJUALAN TERBESAR ---")
print(top5_region)

# 6) Grafik total penjualan berdasarkan region
plt.figure(figsize=(8, 4))
sns.barplot(data=top5_region, x='Region', y='Sales', palette='Greens_r')
plt.title('Total Penjualan Berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
# 7) Tampilkan total Profit berdasarkan Region
profit_region = df.groupby('Region')['Profit'].sum().reset_index()
print("--- TOTAL PROFIT BERDASARKAN REGION ---")
print(profit_region)

# 8) Grafik total profit berdasarkan region
plt.figure(figsize=(8, 4))
sns.barplot(data=profit_region, x='Region', y='Profit', palette='Purples_r')
plt.title('Total Profit Berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

```python
# 8) Perbaikan grafik total profit berdasarkan region
plt.figure(figsize=(12, 6)) # Ukuran grafik diperlebar jadi 12x6

sns.barplot(data=profit_region, x='Region', y='Profit', palette='Purples_r')

plt.title('Total Profit Berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')

# Trik rahasia: putar tulisan nama region miring 45 derajat biar gak tabrakan
plt.xticks(rotation=45, ha='right')

plt.tight_layout() # Biar posisi grafik dan teksnya pas pas dicrop nanti
plt.show()
```

```python
# --- TAHAP 4: ANALISIS PENJUALAN BERDASARKAN PRODUK ---

# 1) Total penjualan berdasarkan Category
print("--- TOTAL PENJUALAN PER KATEGORI ---")
sales_cat = df.groupby('Category')['Sales'].sum().reset_index()
print(sales_cat)
print("\n")

# 2) Visualisasi total penjualan berdasarkan Category
plt.figure(figsize=(6, 4))
sns.barplot(data=sales_cat, x='Category', y='Sales', palette='pastel')
plt.title('Total Penjualan Berdasarkan Kategori Produk')
plt.show()
print("\n")

# 3) Tampilkan 5 Sub-Category dengan total penjualan terbesar
print("--- TOP 5 SUB-CATEGORY PENJUALAN TERBESAR ---")
top5_subcat = df.groupby('Sub-Category')['Sales'].sum().nlargest(5).reset_index()
print(top5_subcat)
```

```python
# 4) Visualisasi 5 Sub-Category penjualan terbesar
plt.figure(figsize=(8, 4))
sns.barplot(data=top5_subcat, x='Sub-Category', y='Sales', palette='magma')
plt.title('Top 5 Sub-Kategori dengan Total Penjualan Terbesar')
plt.show()
print("\n")

# 5) Rata-rata penjualan berdasarkan Category
print("--- RATA-RATA PENJUALAN PER KATEGORI ---")
print(df.groupby('Category')['Sales'].mean())
print("\n")

# 6) Rata-rata penjualan berdasarkan Sub-Category
print("--- RATA-RATA PENJUALAN PER SUB-KATEGORI ---")
print(df.groupby('Sub-Category')['Sales'].mean())
```

```python
# 7) Total Quantity berdasarkan Category
print("--- TOTAL QUANTITY PER KATEGORI ---")
print(df.groupby('Category')['Quantity'].sum())
print("\n")

# 8) Total Quantity berdasarkan Sub-Category
print("--- TOTAL QUANTITY PER SUB-KATEGORI ---")
print(df.groupby('Sub-Category')['Quantity'].sum())
print("\n")

# 9) Total Profit berdasarkan Category
print("--- TOTAL PROFIT PER KATEGORI ---")
print(df.groupby('Category')['Profit'].sum())
print("\n")

# 10) Total Profit berdasarkan Sub-Category
print("--- TOTAL PROFIT PER SUB-KATEGORI ---")
print(df.groupby('Sub-Category')['Profit'].sum())
```

```python
# --- TAHAP 5: VISUALISASI BERDASARKAN VARIABEL KATEGORI ---

# Mengatur tampilan grafik agar rapi dan teksnya tidak tabrakan
plt.rcParams.update({'font.size': 10})

# 1) Grafik Sales berdasarkan Ship Mode
plt.figure(figsize=(6, 4))
df.groupby('Ship Mode')['Sales'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Penjualan Berdasarkan Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()

# 2) Grafik Sales berdasarkan Segment
plt.figure(figsize=(6, 4))
df.groupby('Segment')['Sales'].sum().plot(kind='bar', color='lightgreen')
plt.title('Total Penjualan Berdasarkan Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()

# 3, 4, 5) Grafik Sales berdasarkan Region, Category, dan Order Priority
plt.figure(figsize=(10, 4))
df.groupby('Region')['Sales'].sum().plot(kind='bar', color='coral')
plt.title('Total Penjualan Berdasarkan Region')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(6, 4))
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='gold')
plt.title('Total Penjualan Berdasarkan Kategori')
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(6, 4))
df.groupby('Order Priority')['Sales'].sum().plot(kind='bar', color='orchid')
plt.title('Total Penjualan Berdasarkan Order Priority')
plt.xticks(rotation=0)
plt.show()

# 9) Grafik Profit berdasarkan Order Priority
plt.figure(figsize=(6, 4))
df.groupby('Order Priority')['Profit'].sum().plot(kind='bar', color='salmon')
plt.title('Total Profit Berdasarkan Order Priority')
plt.xticks(rotation=0)
plt.show()
```

```python
# 6) Kombinasi Region dan Category
plt.figure(figsize=(12, 5))
sns.barplot(data=df, x='Region', y='Sales', hue='Category', errorbar=None, palette='Set2')
plt.title('Total Penjualan Berdasarkan Kombinasi Region dan Category')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.show()

# 7) Kombinasi Ship Mode dan Category
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='Ship Mode', y='Sales', hue='Category', errorbar=None, palette='Set1')
plt.title('Total Penjualan Berdasarkan Kombinasi Ship Mode dan Category')
plt.legend(title='Category')
plt.show()

# 8) Kombinasi Segment dan Category
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='Segment', y='Sales', hue='Category', errorbar=None, palette='Pastel1')
plt.title('Total Penjualan Berdasarkan Kombinasi Segment dan Category')
plt.legend(title='Category')
plt.show()
```
